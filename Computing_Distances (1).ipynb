```python
# importing necessary libraries

import mdtraj
import numpy

# import graphics library

%pylab inline
import matplotlib.pyplot as plt
```

```python
traj_1 = mdtraj.load('trj155.lh5')
```

```python
traj_2 = mdtraj.load('trj893.lh5')
```

```python


combined_traj = traj_1 + traj_2
```

```python
combined_traj
```

```python
# computing the distance between pairs of atoms in residues 0 and 10 and residues 0 and 11.
combined_mdtraj_distance = mdtraj.compute_distances(combined_traj,atom_pairs = [[0, 10], [0, 11]])
```

```python
# computing displacement vector between pairs of atom in each frame of a trajectory
combined_mdtraj_dislacement = mdtraj.compute_displacements(combined_traj,atom_pairs = [[0, 10], [0, 11]])
```

```python
combined_mdtraj_contacts = mdtraj.compute_contacts(combined_traj, [[0, 10], [0, 11]])
```

```python
combined_mdtraj_closest_heavy = mdtraj.compute_contacts(combined_traj, [[0, 10], [0, 11]], scheme='closest-heavy')
```

```python
combined_mdtraj_ca = mdtraj.compute_contacts(combined_traj, [[0, 10], [0, 11]], scheme='ca')
```

```python
combined_mdtraj_sidechain = mdtraj.compute_contacts(combined_traj, [[0, 10], [0, 11]], scheme='sidechain')
```

```python
combined_mdtraj_sidechain_heavy = mdtraj.compute_contacts(combined_traj, [[0, 10], [0, 11]], scheme='sidechain-heavy')
```
